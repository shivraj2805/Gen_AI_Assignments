```python
# Gaussian Mixture Model (GMM) - It is a probabilistic generative model that assumes that the data is generated from a mixture of several Gaussian distributions (normal distributions) with unknown parameters.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
# Step 1: Generate sample 2D data (2 clusters)
np.random.seed(0)
mean1 = [0, 0]
cov1 = [[1, 0], [0, 1]]
data1 = np.random.multivariate_normal(mean1, cov1, 100)

mean2 = [5, 5]
cov2 = [[1, 0], [0, 1]]
data2 = np.random.multivariate_normal(mean2, cov2, 100)

X = np.vstack((data1, data2))  # Combine data
```

```python
# Step 2: Fit Gaussian Mixture Model
gmm = GaussianMixture(n_components=2, covariance_type='full', random_state=0)
gmm.fit(X)
```

```python
# Step 3: Generate new samples
new_samples, _ = gmm.sample(5)  # Generate 5 new points
```

```python
# Step 4: Print results
print("Means of Gaussian components:\n", gmm.means_)
print("Covariances of Gaussian components:\n", gmm.covariances_)
print("Generated new samples:\n", new_samples)
```

```python
plt.scatter(X[:,0], X[:,1], label='Original Data')
plt.scatter(new_samples[:,0], new_samples[:,1], color='red', marker='x', label='Generated Samples')
plt.title("Gaussian Mixture Model - Original vs Generated")
plt.legend()
plt.show()
```
