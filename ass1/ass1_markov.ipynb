```python
# Markov Chain -  A Markov Chain predicts the next word based only on the current word, not the full sentence history.
```

```python

import random

class MarkovTextGenerator:

    def __init__(self):
        self.chain = {}

    def train(self, text):
        words = text.split()
        for i in range(len(words) - 1):
            current_word = words[i]
            next_word = words[i + 1]

            if current_word not in self.chain:
                self.chain[current_word] = []
            self.chain[current_word].append(next_word)

    def generate_text(self, start_word, length):
        current_word = start_word
        result = [current_word]

        for _ in range(length):
            if current_word not in self.chain:
                break
            next_word = random.choice(self.chain[current_word])
            result.append(next_word)
            current_word = next_word

        return " ".join(result)
```

```python

# Training the model
model = MarkovTextGenerator()
training_text = "Generative AI models learn probability distributions from data"
model.train(training_text)

# Generating text
generated_text = model.generate_text("Generative", 6)
print(generated_text)
```
